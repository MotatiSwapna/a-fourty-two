```python
import spacy
import textacy
nlp=spacy.load('en_core_web_sm')
```

Opening the file and read the content of the file

```python
f = open('js_nlp.txt','rt')
content = f.read()
```

Dividing the entire content into set of statements

```python
#sentence segmentation
sentences=content.split(".")
print(sentences)
```

Dividing the content into tokens (we can also divide each statement into tokens individually)

```python
#2 tokenization
tokens=nlp(content)
for i in tokens:
  print(i,end='  ')
```

predicting parts of speech of each token

```python
for i in tokens:
  print(i,i.pos_)
```

Text Lemmatization

```python
for i in tokens:
  print(i,i.lemma_)
```

Identifying the stop words

```python

mytokens = []
mywords = []
for token in tokens:
  if(token.is_stop == False):
    mytokens.append(token)
    mywords.append(token.text)
print("Final tokens after removing stop words")
print(mytokens)
print((" ".join(mywords))[:])
```

Dependency Parsing

```python
from spacy import displacy
displacy.render(tokens, style='dep', jupyter=True, options={'distance':100})
```

Finding Noun Phrases

```python
for i in tokens.noun_chunks:
  print(i.text)
```

Named Entity Recognition (NER)

```python
for word in tokens.ents:
    print(word.text,word.label_)
```

Text Summarization

```python
import textacy
statements = textacy.extract.semistructured_statements(tokens,entity='India',cue='be')
# Print the results
print("Some Facts about our country India:")
for statement in statements:
  s, v, f = statement
  print(f" - {f}")
```
